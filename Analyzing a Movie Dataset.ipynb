```python
import pandas as pd

# Read the actual file name 
df = pd.read_csv(r"C:\Users\HP\Desktop\tmdb_5000_movies.csv")

# Preview the first few rows of the dataset
df.head()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Example: Fill missing values with 'Unknown' or a relevant value
df.fillna('Unknown', inplace=True)
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# Convert the 'release_date' column to datetime with error handling
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
```

```python
# Convert the 'release_date' column to datetime with error handling
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Now you can access the year using .dt.year
df['release_year'] = df['release_date'].dt.year

```

```python
# Count the occurrences of each genre
genre_counts = df['genres'].value_counts()
genre_counts.head(10)  # Display the top 10 genres
```

```python
# Sort the dataset by revenue to find the top-grossing movies
top_grossing = df.sort_values(by='revenue', ascending=False).head(10)
top_grossing[['title', 'revenue']]
```

```python
# Calculate the correlation between budget and revenue
correlation = df[['budget', 'revenue']].corr()
print(correlation)
```

```python
# Count the number of movies produced each year
production_trend = df['release_year'].value_counts().sort_index()
production_trend.plot(kind='line')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot genre distribution
plt.figure(figsize=(10,6))
sns.barplot(x=genre_counts.index, y=genre_counts.values)
plt.title('Genre Distribution')
plt.xlabel('Genres')
plt.ylabel('Number of Movies')
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter plot of budget vs. revenue
plt.figure(figsize=(10,6))
sns.scatterplot(x='budget', y='revenue', data=df)
plt.title('Budget vs. Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```

```python
# Line plot of movie production trend over years
plt.figure(figsize=(10,6))
production_trend.plot(kind='line')
plt.title('Movie Production Trend Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.show()
```
